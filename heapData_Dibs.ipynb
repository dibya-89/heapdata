```python
import heapq
import random
#Createing Primary Heap List
list_length = int(input('Enter the length of the list it will create with random number in between 1-100: '))
    # number of elements as input
heap_List=[]
for i in range(0,list_length):
    heap_List.append(random.randint(1,100))
    #heap_List=final_list
# User Choice Taken for Max heap or Min heap
while True:
    choice_Heap= input("Please enter your choice for max= Max Heap or min= Min Heap")
    if choice_Heap == 'max':
        print ("Your Current List which is going to be Max heap", heap_List)
        #Creating max heap form the list
        heapq._heapify_max(heap_List)
        print ("Your Max heap is: ")
        print (heap_List)
        #Taking User input for delete or insert
        operation_Choice=input ("Please choice whiich operation you want to do in Max Heap i= insert and d= delete")
        if operation_Choice == 'i':
            #Taking user input for insert element
            input_Value= int (input ("Enter the value that you want to insert"))
            #Inserting the element in the heap list and heapfiy again
            heapq.heappush(heap_List, input_Value)
            heapq._heapify_max(heap_List)
            #Print the Update Max heap to user
            print("Your Max Heap after Inserting an element is ",heap_List)
        else:
            delete_Value = int(input("Enter the value that you want to delete from the heap"))
            #Search the max value of given list before deleting because it is important to delete the max value only. 
            max_Value= max(heap_List)
            #check the value of max
            if delete_Value == max_Value:
                heapq.heappop(heap_List)
                heapq._heapify_max(heap_List)
                print ("After deleteing your max heap is ", heap_List)
            else:
                print("Deleting is not possible becasue it is not maximum number")
    else:
        print ("Your Current List which is going to be Max heap", heap_List)
        #Creating min heap form the list
        heapq.heapify(heap_List)
        print ("Your min heap is: ")
        print (heap_List)
        #Taking User input for delete or insert
        operation_Choice=input ("Please choice whiich operation you want to do in Min Heap i= insert and d= delete")
        if operation_Choice == 'i':
            #Taking user input for insert element
            input_Value= int (input ("Enter the value that you want to insert"))
            #Inserting the element in the heap list and heapfiy again
            heapq.heappush(heap_List, input_Value)
            heapq.heapify(heap_List)
            #Print the Update Max heap to user
            print("Your min Heap after Inserting an element is ",heap_List)
        else:
            delete_Value = int(input("Enter the value that you want to delete from the heap"))
            #Search the max value of given list before deleting because it is important to delete the min value only. 
            max_Value= min(heap_List)
            #check the value of min with given value
            if delete_Value == max_Value:
                heapq.heappop(heap_List)
                heapq.heapify(heap_List)
                print ("After deleteing your min heap is ", heap_List)
            else:
                print("Deleting is not possible becasue it is not minimum number")
    cont = input("Do you want to continue? yes/no > ")

    while cont.lower() not in ("yes", "no"):
        cont = input("Do you want to continue? yes/no > ")
    if cont == "no":
        print("Bye for now. See you soon.")
        break
    
    
```
